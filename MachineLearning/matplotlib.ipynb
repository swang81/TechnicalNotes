## 1 基本绘图功能

```python
import matplotlib.pyplot as plt
import random

# x,y
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]

# 1.创建画布
plt.figure(figsize=(20,8), dpi = 80)

# 2.绘制折线
plt.plot(x,y_shanghai)

# 3.显示图像
plt.show()
```

## 2.添加自定义x,y刻度 

```python
# 构建x刻度便签
x_ticks_label = ["11 : {} minute".format(i) for i in x]

# 构造y轴刻度
y_ticks = range(40)

# 修改x,y轴的刻度显示
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])
```

```python
import matplotlib.pyplot as plt
import random

# x,y
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]

# 1.创建画布
plt.figure(figsize=(20,8), dpi = 80)

# 2.绘制折线
plt.plot(x,y_shanghai)

# --------------------------------------------------------
# 修改x,y轴的刻度显示
x_ticks_label = ["11 : {} m".format(i) for i in x]
y_ticks = range(40)

plt.xticks(x[::5],x_ticks_label[::5])  #从头到尾间隔5, 不能直接用字符串数组。先用数字，再换成后面的字符串
plt.yticks(y_ticks[::5])
# ---------------------------------------------------
# 添加网格
plt.grid(True,linestyle='--',alpha = 0.5)

# 添加描述信息
plt.xlabel('time')
plt.ylabel('temperature')
plt.title('midnoon 11am -12am')

# savefig要放在show前面
plt.savefig('test.png')

# # 3.显示图像
plt.show()

```
